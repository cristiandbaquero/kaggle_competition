```python
import pandas as pd
```

```python
df = pd.read_csv('../kaggle_competition/prueba_simple/data/predict.csv')
```

```python
dic_cut = { "Ideal": 5,
            "Premium": 4,
            "Very Good":3,
            "Good": 2,
            "Fair":1
          }
```
